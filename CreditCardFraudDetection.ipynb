```python
#Under Sampling
```

```python
# build a sample dataset containing similar distribution of normal transactions and Fradulent transactions
```

```python
legit_sample=legit.sample(492)
```

```python
samp_df=pd.concat([legit_sample,fraud])
```

```python
samp_df.shape
```

```python
samp_df["Class"].value_counts()
```

```python
samp_df.groupby("Class").mean()
```

```python
X=samp_df.drop(columns=["Class"])
y=samp_df["Class"]
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
print(X.shape,Xtrain.shape,Xtest.shape)
```

```python
model1=LogisticRegression()
model1.fit(Xtrain,Ytrain)
```

```python
# prediction on training data
ypred_train=model1.predict(Xtrain)
print(accuracy_score(Ytrain,ypred_train))
```

```python
# prediction on test data
ypred_test=model1.predict(Xtest)
print(accuracy_score(Ytest,ypred_test))
```
